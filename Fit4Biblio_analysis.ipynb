```python
def get_institution(ar, sep='-'):
    """
    @param AbstractRetrieval ar: object retrived from doi
    """
    inst_list = []
    for aut in ar.authors:
        if aut.affiliation:
            for aff in aut.affiliation.split(';'):
                i = AffiliationRetrieval(aff).affiliation_name
                # we need to remove quote
                if AffiliationRetrieval(aff).city:
                    i = i + sep + AffiliationRetrieval(aff).city
                inst_list.append(i)
        
    return list(set(inst_list))

def get_keywords(ar):
    return ar.authkeywords

def create_pair(list_pair, sep='|'):
    return [list_pair[i][0] + sep + list_pair[i][1] for i in range(len(list_pair))]
```

```python
centers_unique = []
centers_pair = {}
keywords_unique = []
keywords_pair = {}
keyword_center_pair = {}
count = 0
for doi in df['DOI'].tolist():
    count += 1
    print(count
         )
    print(doi)
    if str(doi) == 'nan':
        continue
        
    try:
        ab = AbstractRetrieval(doi, view='FULL')
    except:
        continue
    
    if ab.authors:
        inst_l = get_institution(ab)
        keywords = get_keywords(ab)

        # adding centers
        centers_unique += inst_l
        # creating all the possible combinations
        centers_pair_doi = list(combinations(inst_l, 2))

        pair_str = create_pair(centers_pair_doi)
        for p in pair_str:
            if p in centers_pair.keys():
                centers_pair[p] = centers_pair[p] + 1
            else:
                centers_pair[p] = 1

        # keywords (sometimes they are null)
        if keywords:
            keywords_unique += keywords

            key_pair = create_pair(list(combinations(keywords, 2)))
            for key in key_pair:
                if key in keywords_pair.keys():
                    keywords_pair[key] = keywords_pair[key] + 1
                else:
                    keywords_pair[key] = 1
            # coupling keywords and centers
            for k in keywords:
                for c in list(set(inst_l)):
                    key_kc = k+'|'+c
                    if key_kc in keyword_center_pair.keys():
                        keyword_center_pair[key_kc] = keyword_center_pair[key_kc]+1
                    else:
                        keyword_center_pair[key_kc] = 1
    
centers_unique = list(set(centers_unique))
keywords_unique  = list(set(keywords_unique))
```

## Create the .net format
Reference: https://gephi.org/users/supported-graph-formats/pajek-net-format/

### 1) Nodes file
<code>
*Vertices 82670
1 "entity"
2 "thing"
3 "anything"
4 "something"
5 "nothing"
6 "whole"
    </code>

```python
def write_net_nodes(list_nodes, fpath, sep=' ', header=''):
    list_nodes_u = list(set(list_nodes))
    n = len(list_nodes_u)
    vertices_dict = {}
    with open(fpath, 'w') as o:
        o.write(header+'\n')
        for i in range(len(list_nodes_u)):
            vertices_dict[list_nodes_u[i]] = i+1 #list_nodes_u[i]
            o.write(str(i+1)+sep+list_nodes_u[i].replace("\"",'')+'\n')
    return vertices_dict
            
```

### 2) Edges file
First two columns represent nodes and the last column represent the weight
<code>
    *arcs
4244 107 2
4244 238 3
    
 </code>

```python
def write_edge_file(mdict, fpath, ver_dict, sep=' ', header=''):
    """
    mdict: dict with key as links (sep='|') and value number of occurrence
    ver_dict: dict with mapping between name of the node and its corresponding number in the net node file o
    ver_2: dict with
    """
    with open(fpath, 'w') as o:
        o.write(header)
        for k, v in mdict.items():
            nodes = k.split('|')
            o.write(str(ver_dict[nodes[0]])+sep+str(ver_dict[nodes[1]])+sep+str(v)+'\n')
```

```python
kw_center_path = '/Users/giovannanicora/Desktop/fit4biblio/keywords_centers.txt'
center_path = '/Users/giovannanicora/Desktop/fit4biblio/centers_links.txt'
gephi_path = '/Users/giovannanicora/Desktop/fit4biblio/gephi_keywords_centers.csv'
gephi_center_path = '/Users/giovannanicora/Desktop/fit4biblio/gephi_centers.csv'

# mettiamo tutto in lower case
def write_gephi_csv_from_txt(fpath_list, opath, sep=','):

    with open(opath, 'w') as o:
        for fpath in fpath_list:
            with open(fpath, 'r') as f:
                next(f)
                for line in f:
                    lsp = line.split('\t')
                    n = int(lsp[-1].strip())
                    for i in range(n):
                        o.write(lsp[0].lower()+sep+lsp[1].lower()+'\n')

                
```
